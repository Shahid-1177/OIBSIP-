MODEL TRAINING

```python
from sklearn.model_selection import train_test_split
# train - 70
# test - 30
X = iris.drop(columns=['Species'])
Y = iris['Species']
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.30)
```

```python
# logistic regression
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train, y_train)
print("Accuracy: ",model.score(x_test, y_test) * 100)
```

```python
# knn - k-nearest neighbours
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
model.fit(x_train, y_train)
print("Accuracy: ",model.score(x_test, y_test) * 100)
```

```python
# decision tree
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
print("Accuracy: ",model.score(x_test, y_test) * 100)
```

```python
# random forest
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(x_train, y_train)
print("Accuracy: ",model.score(x_test, y_test) * 100)
```

```python
# svm
from sklearn.svm import SVC
model = SVC()
model.fit(x_train, y_train)
print("Accuracy: ",model.score(x_test, y_test) * 100)
```

**Withou Outlayers From Sepalwidth**


```python
# Remove outliers from 'SepalWidthCm' using IQR
Q1 = iris['SepalWidthCm'].quantile(0.25)
Q3 = iris['SepalWidthCm'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

iris_no_outliers = iris[(iris['SepalWidthCm'] >= lower_bound) & (iris['SepalWidthCm'] <= upper_bound)]

print("Original DataFrame shape:", iris.shape)
print("DataFrame shape after removing outliers:", iris_no_outliers.shape)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
iris['Species'] = le.fit_transform(iris['Species'])
iris.head()
```

```python
from sklearn.model_selection import train_test_split
# train - 70
# test - 30
X = iris.drop(columns=['Species'])
Y = iris['Species']
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.30)
```

```python
# svm
from sklearn.svm import SVC
model = SVC()
model.fit(x_train, y_train)
print("Accuracy: ",model.score(x_test, y_test) * 100)
```

```python
# logistic regression
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train, y_train)
print("Accuracy: ",model.score(x_test, y_test) * 100)
```

```python
# knn - k-nearest neighbours
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
model.fit(x_train, y_train)
print("Accuracy: ",model.score(x_test, y_test) * 100)
```

```python
from sklearn.metrics import classification_report

# Classification report for KNN
print("Classification Report for KNN:")
knn_model = KNeighborsClassifier()
knn_model.fit(x_train, y_train)
y_pred_knn = knn_model.predict(x_test)
print(classification_report(y_test, y_pred_knn))
```

```python
from sklearn.metrics import classification_report

# Classification report for Logistic Regression
print("Classification Report for Logistic Regression:")
lr_model = LogisticRegression()
lr_model.fit(x_train, y_train)
y_pred_lr = lr_model.predict(x_test)
print(classification_report(y_test, y_pred_lr))
```

```python
from sklearn.metrics import classification_report

# Classification report for SVM
print("Classification Report for SVM:")
svm_model = SVC()
svm_model.fit(x_train, y_train)
y_pred_svm = svm_model.predict(x_test)
print(classification_report(y_test, y_pred_svm))
```
